# Magic Number
This code manages the allocation of mentees to mentors in IMP, ensuring that no mentor has too many mentees. TheMagicNumber is a variable that sets the maximum number of mentees that a mentor can have across all waves of the program. For example, if TheMagicNumber is set to 4 and we are currently in wave 3, each mentor can have a maximum of 4 mentees across all three waves combined. This helps to ensure that each mentor has a fair and manageable workload. By editing TheMagicNumber, you can adjust the maximum number of mentees per mentor and customize the program to fit your needs.

```python
#Can EDIT
TheMagicNumber = 2
```

The following cell will run the magic mentoring machine and spit out three different outputs. 
1. The schedule variable will represent what our schedule will look like once we've matched all of the students in our current wave.
2. The extras variable will represent all students that we weren't able to schedule for this current wave.
3. The mentorCapacity variable will contain all mentors and the number of students they have collected in pairs.


```python
schedule, extras, mentorCapacity = met.everything(TheMagicNumber)
```
